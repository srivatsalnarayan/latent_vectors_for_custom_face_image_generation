```python
import cv2
import os
import glob

# Specify the path to your image folder
folder_path = 'crop_part1'

# Initialize img_number as a global variable
img_number = 0

# Load the Haar Cascade Classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Create an "output_faces" folder if it doesn't exist
output_folder = 'output_faces'
os.makedirs(output_folder, exist_ok=True)

# Function to detect and save faces from an image
def detect_and_save_faces(image_path):
    global img_number  # Declare img_number as global
    
    # Load the image
    img = cv2.imread(image_path, cv2.IMREAD_COLOR)
    
    # Check if the image was loaded successfully
    if img is not None:
        # Convert the image to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        # Detect faces using the loaded cascade classifier
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
        
        try:
            for (x, y, w, h) in faces:
                # Crop the detected face
                roi_color = img[y:y+h, x:x+w]
                
                # Resize the cropped face to 128x128 pixels
                resized = cv2.resize(roi_color, (128, 128))
                
                # Save the cropped and resized face to the output folder
                output_path = os.path.join(output_folder, f"face_{img_number}.jpg")
                cv2.imwrite(output_path, resized)
                
                # Increment the image number
                img_number += 1
        
        except:
            print(f"No faces detected in {image_path}")

# Use glob to get a list of image files in the folder
image_files = glob.glob(os.path.join(folder_path, '*.jpg')) + glob.glob(os.path.join(folder_path, '*.png'))

# Process each image
for file in image_files:
    print(f"Processing: {file}")
    detect_and_save_faces(file)

print(f"Total {img_number} faces detected and saved.")
```
